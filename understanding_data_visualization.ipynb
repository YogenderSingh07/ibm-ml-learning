```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('/content/mckinsey (1).csv')
print(data)
```

```python
x_bar = data['continent'].value_counts().index
y_bar = data['continent'].value_counts()
plt.bar(x_bar,y_bar,color='pink')
```

```python

plt.figure(figsize=(8,6))
plt.xticks(rotation = 30)
plt.xlabel('Continents',color ='Green',fontsize=12)
plt.ylabel('Count of continents',color ='Green',fontsize=12)
plt.title('Countries and their Count')
plt.bar(x_bar,y_bar,color='pink')
```

```python
plt.hist(data['life_exp'],bins = 12)
```

```python
sns.countplot(data = data, x ='continent',color ='pink')
plt.xticks(rotation = 45)
```

```python
sns.histplot(data = data, x='life_exp',bins =10)
```

```python
sns.histplot(data = data, x='life_exp',hue ='continent')
```

```python
sns.kdeplot(data['life_exp'],shade=True)
```

```python
#want to find the aggregates like median, min ,max ,percentiles of the data
# we use boxplot
  # lower whisker
  #first quartile
  # median
  # second quartiles
  # max
# those circles are outliers
sns.boxplot(data=data,y="gdp_cap",x = "continent")

```

```python
##plotting gdp_cap of highest gdp_cap country over the years


#finding the highest gdp_per_cap country
x = data.sort_values('gdp_cap',ascending=False)

##putting all the data of kuwait in a new dataframe
kuwait =data.loc[data['country']=='Kuwait']



```

```python
x= data.sort_values('gdp_cap',ascending=False)
x.head(10)
##finding data for norway

norway = data.loc[data['country']=='Norway']
```

```python
Japan= data.loc[data['country']=='Japan']
```

```python
## to compare two numerical features
## use lineplot -- trend yearwise

##setting our dataset as the data of kuwait and then plotting the gdp of kuwait over the years
sns.lineplot(data=kuwait,x='year',y='gdp_cap',color='Purple')
sns.lineplot(data=norway,x='year',y='gdp_cap',color='blue')
sns.lineplot(data=Japan,x='year',y='gdp_cap',color='Green')
plt.legend(labels =['Kuwait','norway','Japan'])
```

```python
#SCATTER PLOT  --used for negative correlation , higher rank for sales of game means lower sales
sns.scatterplot(data=kuwait,x='year',y='life_exp')
sns.scatterplot(data=norway,x='year',y='life_exp')
sns.scatterplot(data=Japan,x='year',y='life_exp')
```

```python
##
sns.scatterplot(data=data,x='year',y='life_exp', hue='continent')
plt.show()
```

```python
##Compare top3 continent(frequency) with 3 countries(),(categorical vs categorical)

dx = data['continent'].value_counts()
top3data = data.loc[data['continent'].isin(['Africa','Asia','Europe'])]

#top3 of every continent
top3_africa = top3data.loc[top3data['continent']=='Africa'].head(3)
top3_asia = top3data.loc[top3data['continent']=='Asia'].head(3)
top3_europe = top3data.loc[top3data['continent']=='Europe'].head(3)

##
top3_countries = top3data.loc[top3data['country'].isin(['Algeria','Afghanistan','Albania'])]

sns.countplot(data=top3_countries, x='continent',hue='country')
```

```python
##categorical vs numerical can be represented in both boxplot and barplot

#Q - continent vs gdp_cap

data['continent'].value_counts()
top3_cont = data.loc[data['continent'].isin(['Africa','Asia','Europe'])]


sns.boxplot(data = top3_cont, x ='continent', y='gdp_cap')
```
