```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/content/football_data.csv')
print(df)
```

```python
##Displaying first 5 and last rows
print(df.head())
print(df.tail())
```

```python
##Shape of the dataset
print(df.shape)
```

```python
##List all column names and their datatypes

print(df.info())
```

```python
# find if columns have null values
print(df.isna().sum())
```

```python
#Verify that home_possession + away_possession = 100 for every match

is_total_possession_100 = (df['home_possession'] + df['away_possession']==100).all()
print(is_total_possession_100)
```

```python
# create new col total_goals
df['Total_Goals'] = df['home_goals'] + df['away_goals']
df
```

```python
#create new Col Goal_Difference
df['Goal_Difference'] = df['home_goals']-df['away_goals']
```

```python
##Add a new col that shows the match result -- if else conditon cant be directly applied on df , it has to be applied on a single row
def match_result(row):
  if row['home_goals'] > row['away_goals']:
    return 'Home Win'
  elif row['home_goals']==row['away_goals']:
    return 'Draw'
  else :
    return 'Away Win'

df['Match_result'] = df.apply(match_result,axis=1)
print(df)
```

```python
# Create a boolean column high_scoring_match (total goals  4)
df['high_scoring_match'] = df['Total_Goals']>=4
print(df)
```

```python
#Extract year and month from the date
df['year']= df['date'].str.split("-").str[0]
df['month']=df['date'].str.split("-").str[1]
df
```

```python
#Create a column more_shots_team (Home / Away / Equal)
def more_Shots(row):
  if row['home_shots']>row['away_shots']:
    return "Home"
  elif row['home_shots']==row['away_shots']:
    return "Draw"
  else:
    return "Away"
df['more_shots_team'] = df.apply(more_Shots,axis=1)
df
```

                          FILTERING AND SELECTION

```python
#Find all matches where the home team scored more than 2 goals
dx = df.loc[df['home_goals']>2]
dx
```

```python
#List matches where possession difference is greater than 20%
dn = df.loc[abs(df['home_possession']-df['away_possession']>20)]
dn
```

```python
#Select only Premier League matches
only_prem_matches = df.loc[df['league']=='Premier League']
only_prem_matches
```

```python
#Find matches played after September 2023 -- two ways,convert date string to date time else convert year and month col to int

df['month'] = df['month'].astype(int)
df['year'] = df['year'].astype(int)

matches_after_sep_23 = df.loc[(df['month']>9) & (df['year']>=2023)]
matches_after_sep_23
```
